```python
import numpy as np
import pandas as pd
```

```python
def abcd(file_name):
    file = open(file_name,'r')
    data = file.read()
    ch = ""
    ch2 = "Body$Thread_Id$Date$Date_Sent$Address$\n"
    #print(data)
    for x,y in enumerate(data):
    #    print(x)
        if y=='\n':
    #        print(x,y)
            ch += " "
        elif y==',' and data[x-1]=='$':
            ch += "\n"
        else:
            ch += y
    #print(ch)
    file2 = open('myfile3.txt','w')
    file2.write(ch2)
    file2.write(ch)
    file.close()
    file2.close()
```

```python
abcd('myfile.txt')
```

```python
train = pd.read_csv('myfile3.txt',delimiter='$')
```
